```python
#Distribution (the bilue line is a distribution line)
vis1 = sns.distplot(df["InternetUsers"]) # UNIVARIATE ANALYSIS - STATISTICS
# we plot this graph using one variable called internetusers so that it is called univariable
# as we see in excel sheet internet users starting point is 0.9 and ending point is 98 so that likewise iin graph also 
# by seeing the graph we can say that 0 to 9 having max freq mean max internet users amd likewise min users also
# if instead of sns we use plt it will show error bcz sns we use for advance visualization
```

plot the graph using 1 variable is called univariate analysis
plot the graph using 2 variable is called bivariate analysis
plot the graph using >2 variable is called multiivariate analysis

```python
#BOX PLOTS:
vis4 = sns.boxplot(data = df, x="IncomeGroup",y='BirthRate') # bivariant
# in the next session means 8-8-24 sir explained like as the income low point is 7.9 from excel sheet so that the lowest line of 1st box of high income is of 7.9 and the max is 35 so the line should be above 30 but it is not like that the line is on 20 why bcz as we can see 7.9 is near to 8 likewise all the values are near to another but 20 is near to 21 and 35 is not near to 35 so this concept is called outlier 
# that is outlier is the concept which is very far from other observation (sirs defination)
# e.g there are two groups one is boys and second is girls if one girl enter in a boys grp and one boy enter in girls grp then they are outlier bcz they are diff from other
# these oulierss are very powerful in ml they having big impact on algorithm so diff algorithm are introduced to adjust the outliers
```
